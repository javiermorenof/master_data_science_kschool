## Aprendizaje supervisado vs no supervisado

En el aprendeizaje supervisado tenemos etiquetas (y), datos reales, y vamos a querer maximizar (accuracy, recall...) o minimizar (MAE, MSE, Etc.) la predicción.

En ecambio en el aprendizaje no supervisado no tenemos etiquetas así que no podemos comparar nuestros resultados con un training set, pero si hay una serie de técncias que nos permiten comparar algoritmos para tener una idea de si estoy yendo bien o no.

Aplicaciones del aprendizaje no supervisado:
- Segmentacion o clustering. Agrupar individuos en semejantes
- Social Network Analysis. Análisis de redes sociales y grafos.
- Deteccion de outlier. Nosotros tenemos unos datos y queremos saber qué son animalías, como por ejemplo detección de operaciones fraudulentas.
- Reglas de asociación. Aquí encontramos el market basket analysis. Es el precursor de los sistemas de recomendación.

## Kmeans
Tiene dos pasos, el primero es recalcular los centroides. El centroide es el punto medio de todos los puntos del clúster. El siguiente paso es asignar cada punto al centroide más cercano
Posteriormente se sigue iterando y se recalculan los centroides hasta que deja de haber cambios.
El resultado es bastante sensible a dónde se han iniciado los centroides. Esto se soluciona realizando el clustering varias veces y asignando cada punto al clúster en el que más veces ha caído.

Para saber el número de clusters se utiliza generalmente el método del codo, que calcula la distancia de cada punto al centroide y optimiza.
Cuantos más clústers tenga, el valor de J de kmeans va a ser más pequeño.

La principal limitación de Kmeans es que está asumiendo que la densidad es esférica, que no le  va bien identificar clusters de diferentes tamaños ni clusters de diferentes densidades y que es muy sensible a outliers.

Kmeans es muy sensible a la escala de las variables porque se están midiendo distancias.

# Hierarchichal clustering
Tiene dos enfoques: aglomerativo y divisivo..
- Aglomerativo. Considera que N puntos tiene N ckusters. y va combinando los dos puntos que estén más cerca. Y los combina hasta que todos están en el mismo punto.
- Divisivo. Es a al inversa. Indica que todos los puntos están en el mismo cluster y lo va divdiendo.
Requiere una **métrica** que sirva de criterio: el centroide de uno, la distancia máxima entre los puntos, etc.

Se representa con un dendograma, pero nosotros tenemos que decidir luego el nñumero de clusters.

Ventajas:
- No necesitamos que especificar al principio el numero de clusters
- Eso significa que no tenemos que hace asunciones sorbe las dimesniones de los datos.

Desventajas:
- Es necesarios especificar la distancia de threshold
- Es sensible a la distancia y al criterioe de enlace

## DBscan
Una de las ventajas que aporta es que considera que hay puntos que son outliers, cosa que no hacia k-means y el hierarchical clustering.
Detecta diferentes densidades, diferente tamaño, etc.. Tiene difretnes parémetros:
- MinPoints. CUántos puntos tengo que tener para ser considerado un outlier.
- Eps. Distancia entre los puntos.
Ventajas:
- El numero de clusters se define automaticamente.
Desventajas:
- Hay que definir el numero de puntos minimo para un clusters y la distancia.

Como se calcula el eps y el minimo de puntos.
Damos el eps nosotros.

# Implementation of K-Means

In this section we will implement kmeans algorithm. 


## Algorithm

```
clustering (data, K):
    Randomly initialize K cluster centroids (mu(1),..., mu(k))
    # or select K random points from data
    Repeat until convergence:
        # assign cluster
        for d in data:
            assign d to closest cluster centroid
        # recompute cluster centroid
        for k = 1 to K:
            mu(k) = mean of data points assigned to cluster k
            
```

### Exercise 1: complete the following code so that it follows k-means algorithm

- Assume data is a a (Npoints, dim) numpy array. 
- Hint: use `numpy.linalg.norm()` to compute euclidean distance
- Hint: `np.random.choice()` function may be useful

*15 min*

````python

def kmeans (data, K, plot=False):
    N = data.shape[0]
    dim = data.shape[1]
    
    # generate K random centroids
    centroids = data[np.random.choice(N, K, replace=False)] # To-DO
    
    # initialize vectors
    new_centroids = np.zeros((K, dim))
    distances = np.zeros((N, K))
    
    # repeat until convergence
    niter = 1
    while True:
        
        # compute distance of each point to cluster centroid
        for i in range(K):
            distances[:,i] = np.linalg.norm(data-centroids[i], axis=1) #TO-DO
         
        
        # assign points to closest centroid
        clusters = np.argmin(distances, axis=1) #TO-DO
        
        # recompute clusters' centroids
        # Calculate mean for every cluster and update the center
        for i in range(K):
            new_centroids[i] = np.mean(data[clusters == i], axis=0) #TO-DO
        
        # compute if there is any variation
        if np.array_equal(centroids, new_centroids):
            break
            
        centroids = new_centroids.copy()

        if plot:
            # solution
            plt.scatter(data[:,0], data[:,1], c=clusters, cmap='plasma', linewidths=0);
            for k in range(K):
                plt.scatter(centroids[k,0], centroids[k, 1], s=100, marker='D', color='red')
            plt.show()

        niter+=1
        
    return clusters, centroids
````

````python
# let's try our algorithm
K = 3
clusters, centroids = kmeans(data, K)
````

````python
#Pinta como componente de x la primera columna, como componente de Y la segunda columna y con c da el color
plt.scatter(data[:,0], data[:,1], c=clusters, cmap="plasma", linewidths=0);
````

## Computing how good the cluster partition is

Remember SSE (Sum of Squared Error):
$$
SSE = \sum_{i=1}^N (x_i - C_{(X_i)})^2
$$
where $C_{(X_i)}$ represents the cluster centroid of $X_i$.



### Exercise 3: complete SSE function and compute clustering metrics for different number of clusters

SSE calcula la suma de todos los elementos a su centroide

````python
def sse(data, clusters, centroids):
    return np.sum(np.square(np.linalg.norm(data-centroids[clusters], axis=1)))
````

````python
ks = range(1, 20)
sse_errors = np.zeros(len(ks))

for i, k in enumerate(ks):
    clusters_, centroids_ = kmeans(data, k)
    sse_errors[i] = sse(data, clusters_, centroids_)
    print(k, sse_errors[i])
    
    #if np.isnan(sse_errors[i]):
    #    print(clusters_)
    #    print(centroids_)
````

````python
plt.plot(ks, sse_errors, 'o-');
plt.xlim(min(ks), max(ks));
````

# scikit-learn K-means

Now we are going to use scikit-learn library for our exercise

KMeans works as other models in sklearn:

- define the model (and parameters)
- fit the model on training dataset
- apply the fitted model on another dataset (can be the same dataset)

````python
from sklearn.cluster import KMeans
````

````python
# inspect help
?KMeans
````

````python
# define the model
model = KMeans(n_clusters = 3)
````

````python
# fit the model on training data
model = model.fit(data)
````

````python
# press tab to see available methods
#model. #press tab
print(model.cluster_centers_)
print(model.labels_)
print(model.inertia_) # sse
````

````python
# check inertia is our sse defined function
abs(model.inertia_ - sse(data, model.labels_, model.cluster_centers_)) < 0.01
````

````python
# apply the fitted model 
# if applied to the same data, we get model.labels_
# Básicamente el predict gairda los centroides
clusters_sk = model.predict(data)
all(model.labels_ == clusters_sk)
````

````python
clusters_sk
````

````python
def plot_clustering(data, clusters, centroids = []):
    
    K_ = len(set(clusters))
    
    plt.figure(figsize=(8,4))
    plt.scatter(data[:,0], data[:,1], c=clusters, cmap="plasma", linewidths=0)

    if centroids != []:
        for k in range(K_):
            plt.scatter(centroids[k,0], centroids[k, 1], s=100, marker='D', color='red')
        
    plt.show()
````

````python
plot_clustering(data, clusters_sk)
````

In this section we will learn how to apply Hierarchical Clustering in Python. 

We will use `scipy` package:

- `linkage` function computes the distance matrix between the points
- `dendrogram` function plots the dendrogram using the distances

````python
Z = linkage(data, 'ward') #'single', 'ward', ...
````

Now let's understand how the algorithm works. 

It follows an agglomerative approach, so *close* points are merged. 

`linkage` returns how points are merged at each iteration. The output is: [`id_node_1`, `id_node_2`, `distance`, `number_of_points_in_group`]

# New data : non-convex datasets

Now we will apply DBscan on non convex data to see the differences with k-means. We will also learn how to load already-predefined datasets from `sklearn`

````python
from sklearn import datasets

nsamples = 1000
````

````python
# make_blobs
# this creates 'centers' circles
# we will change the covariance so that the clusters become ellipses

X, y = datasets.make_blobs(random_state=170, n_samples=nsamples, centers = 5)
transformation = np.random.RandomState(74).normal(size=(2, 2))
X = np.dot(X, transformation)

# plot
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
````

Compare DBscan against Kmeans on that data

````python
dbscan_model = DBSCAN(eps=0.4, min_samples=5)
dbscan_model = dbscan_model.fit(X)

kmeans_model = KMeans(n_clusters = 5)
kmeans_model = kmeans_model.fit(X)

plot_clustering(X, dbscan_model.labels_)
plot_clustering(X, kmeans_model.labels_, kmeans_model.cluster_centers_)
````
