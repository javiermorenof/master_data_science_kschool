```python
import wget
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/rezpe/datos_viz/master/brain.csv"

wget.download(url)
```

```python
brain = pd.read_csv("brain.csv")
```

```python
brain
```

```python
import altair as alt
```

```python
hist_brain = alt.Chart(brain).mark_bar().encode(
x=alt.X('Brain Weight',bin=alt.Bin(maxbins=100)),
y="count()"
).properties(
width=300,
height=150,
title="Relación peso del cerebro y del cuerpo"
).interactive()

hist_body = alt.Chart(brain).mark_bar().encode(
x=alt.X('Body Weight',bin=alt.Bin(maxbins=100)),
y="count()"
).properties(
width=300,
height=150,
title="Relación peso del cerebro y del cuerpo"
).interactive()

scatter_brain_body = alt.Chart(brain).mark_circle().encode(
x='Body Weight',
y='Brain Weight'
).properties(
width=700,
height=300,
title="Relación peso del cerebro y del cuerpo"
).interactive()


#La barra y entre paréntesis para ponerlo en una misma línea y el & para la línea siguiente
comp_brain = (hist_brain|hist_body)&scatter_brain_body
comp_brain
```

```python
select_term = alt.selection(type="single", encodings = ["x"]) #Encodings sirve para decir qué se filtra

trend_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y="mean(value)",
    color="search_term"
).transform_filter( #transform_filter es el filtrado
    select_term
)

trend_bar = alt.Chart(trends).mark_bar().encode(
    x="search_term",
    y="value",
    color="search_term",
    tooltip="value" #tooltip sirve para que qparezca un cartelito si dejas el ratón por encima
).properties(
    selection=select_term #Y con selection le dices que quieres que filtre este gráfico en base a lo indicado en select_terms
)

trend_bar|trend_line
```

```python
select_date = alt.selection(type="interval",encodings=["x"]) #con interval puedes filtrar por intervalos arrastrando

trend_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y="mean(value)",
    color="search_term"
).properties(
    selection=select_date
)

trend_bar = alt.Chart(trends).mark_bar().encode(
    x="search_term",
    y="value",
    color="search_term",
    tooltip="search_term"
).transform_filter(
    select_date
)

trend_bar|trend_line
```

```python
#Graph to filter and make bigger
#En select_date = alt.selection(type="interval",encodings=["x"])

mini_trend_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y="mean(value)",
    color="search_term"
).properties(
    height=100
).add_selection(
    select_date
)

trend_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y="mean(value)",
    color="search_term"
).transform_filter(
    select_date
)

mini_trend_line&trend_line
```

```python
url4 = "https://raw.githubusercontent.com/rezpe/datos_viz/master/worldTemperatures.csv"
wget.download(url4)
world_temp = pd.read_csv("https://raw.githubusercontent.com/rezpe/datos_viz/master/worldTemperatures.csv",header=None)
```

```python
world_temp.head()
```

```python
world_temp.columns=['0', '1', '2', 'date', '4', 'temp', '6', '7', '8', '9', '10', 'lat', 'lon',
       'city', 'country']


temp = world_temp.groupby("country").mean().reset_index()

alt.Chart(temp).mark_bar().encode(
    x=alt.X("country", sort = alt.Sort(field="temp",op="mean",order="descending")), #con el sort ordeno, pero tento que indicar antes que uso altair y que estoy en la X (alt.X)
    y="mean(temp)"
)
```

```python
world_temp["date"]=pd.to_datetime(world_temp["date"],format="%Y%m%d")

temp = world_temp.groupby("date").mean().reset_index()

alt.Chart(temp).mark_line().encode(
    x=alt.X("date:T"),
    y="mean(temp)",
).properties(
    width=600
)
```

```python
select_country = alt.selection(type="single",encodings=["x"])

temp = world_temp.groupby(["country","date"]).mean().reset_index()

temp_bar = alt.Chart(temp).mark_bar().encode(
    x=alt.X("country",sort=alt.Sort(field="temp",op="mean",order="descending"),axis=None),
    y="mean(temp)",
    tooltip="country"
).properties(
    width=600
).add_selection(select_country)

temp_line = alt.Chart(temp).mark_line().encode(
    x=alt.X("date:T"),
    y="mean(temp)",
).properties(
    width=600
).transform_filter(
    select_country
)

temp_bar&temp_line
```
