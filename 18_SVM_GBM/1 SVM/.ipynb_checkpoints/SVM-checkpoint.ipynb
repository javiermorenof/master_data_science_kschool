# Maximum Margin Classifiers
<div style = "float:right"><a style="text-decoration:none" href = "#inicio">Inicio</a></div>

SVM tiene un enfoque antiguo (creado en los 60's) y que tiene algunas limitaciones prácticas.
SVM fue creadas en entornos de clasificación, aunque posteriormente se utilizó en entornos de regresión, aunque lo más útil y naturales es en entornos de clasificación.

Se basa en los clasificadores de máximo margen, que busca crear un hiperplano que diste lo máximo posible de los bloques que se quieren separar.
Llamamos **margen** a la suma de la distancias entre los elementos más próximos de ambos grupos.
Esos elementos más cercanos para cada grupo que establecen el margen se llaman **vectores de soporte**.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = np.load("XClas.npy")
y = np.load("yClas.npy")
```

```python
from sklearn.svm import SVC
```

### La función de decisión
<div style = "float:right"><a style="text-decoration:none" href = "#inicio">Inicio</a></div>

Los clasificadores realmente no devuelve 0 o 1, sino un valor y a ese valor se le asigna una clase. Mediante ***decision_function*** obtenemos este valor.
En este caso tiene pinta que todos los valores negativos van a recibir un valor 0 y todos los positivos van a recibir un valor 1

```python
sample = np.array([3, -1]).reshape(1, -1)
model.decision_function(sample)
```

```python
fig, ax = draw(X, y, show = False)
ax.scatter([3], [-1], c = "Red")
plt.show()
```

###### Calculemos la función de decisión para otros puntos

###### Función de decisión para una matriz de puntos

```python
plt.figure(figsize = (10, 8))
plt.scatter(X[:, 0], X[:, 1])
for cx in range(-3, 3):
    for cy in range(-2, 3):
        sample = np.array([cx, cy]).reshape(1, -1)
        value = model.decision_function(sample)[0]
        plt.text(cx, cy, round(value, 2), fontsize = 10)
plt.ylim(-2.5, 3.5)
plt.xlim(-3.5, 3.5)
plt.show()
```

```python
colors = ["salmon", "lightgreen"]
plt.figure(figsize = (10, 8))
for cx in np.linspace(-3, 2, 25):
    for cy in np.linspace(-2, 2, 25):
        sample = np.array([cx, cy]).reshape(1, -1)
        value = model.predict(sample)[0]
        plt.scatter(cx, cy, c = colors[value])
plt.show()
```

Sin embargo, el mundo real es más complicado y debemos admitir cierto margen de error, ya que las divisiones no son tan claras.

Esto se regula con el parámetro de regularización C. Cuando la C es grande tenemos un modelo de baja regularización, donde los márgenes son muy estrechos. Cuando la C es pequeña, tenemos un modelo de gran regulariación. Debemos evitar modelos con márgenes muy muy pequeños porque queremos un margen de error.

El error es la distancia entre la muestra mal clasificada y el límite del margen en el qjue debería estar

```python
X = np.load("XSMClas.npy")
y = np.load("ySMClas.npy")
```

###### Instanciamos SVC con un valor de C = 1 y un kernel lineal

```python
#Con C el margen será relativamente grande
model = SVC(kernel = 'linear', C=1)

model.fit(X,y)
```

```python
draw(X, y, model)
```

# Kernels
<div style = "float:right"><a style="text-decoration:none" href = "#inicio">Inicio</a></div>

<div style="background:#0099cc;padding:20px;color:#ffffff">
<strong>EJERCICIO</strong>

</div>

```python
iris = pd.read_csv("iris.csv")
iris.head()
iris = iris.reindex(["sepal_length", "sepal_width", "species"], axis = 1)
iris = iris.sample(frac = 1)
y = iris.pop("species").astype("category").cat.codes.values
X = iris.values
```

```python
draw(X, y);
```

### Linear kernel

```python
model = SVC(kernel = "linear")
model.fit(X, y)
draw3(X, y, model);
```

### Polynomial kernel

```python
model = SVC(kernel = "poly", degree = 1, gamma = 1, coef0 = 0)
model.fit(X, y)
draw3(X, y, model);
```

```python
model = SVC(kernel = "poly", degree = 2)
model.fit(X, y)
draw3(X, y, model);
```

```python
model = SVC(kernel = "poly", degree = 3)
model.fit(X, y)
draw3(X, y, model);
```

```python
model = SVC(kernel = "poly", degree = 3, C = 10)
model.fit(X, y)
draw3(X, y, model);
```

### Radial Basis Function (Gaussian)

Si se asigna a &gamma; el valor "auto", tomará el valor 1/n, siendo n el número de características.

Si gamma es pequeño tiende a un comportamiento lineal, pero si es grande tiende al sobrentrenamiento.

```python
model = SVC(kernel = "rbf", gamma = "auto")
model.fit(X, y)
draw3(X, y, model);
```

Si &gamma; es muy pequeño, la función tenderá a comportarse linealmente:

```python
model = SVC(kernel = "rbf", gamma = 0.01)
model.fit(X, y)
draw3(X, y, model);
```

Y si es muy grande, tenderá al sobreentrenamiento:

```python
model = SVC(kernel = "rbf", gamma = 10)
model.fit(X, y)
draw3(X, y, model);
```

### Sigmoid kernel



Se le llama también perceptrón neuronal

```python
gammas = [0.001, 0.01, 0.1, 1]
for g in gammas:
    print(g)
    model = SVC(kernel = "sigmoid", gamma = g, coef0 = 0)
    model.fit(X, y)
    draw3(X, y, model);
```

# Kernels personalizados
<div style = "float:right"><a style="text-decoration:none" href = "#inicio">Inicio</a></div>

```python
iris = sns.load_dataset("iris")
iris = iris[iris.species.isin(["setosa", "virginica"])]
iris = iris.loc[:, ["sepal_length", "sepal_width", "species"]]
iris = iris.sample(frac = 1)
y_iris = iris.pop("species").astype("category").cat.codes.values
X_iris = iris.values
```

```python
draw(X_iris, y_iris);
```

```python
def custom_kernel(X, Y):
    M = np.array([[2, 0], [0, 1]])
    return np.dot(np.dot(X, M), Y.T)
```

```python
model = SVC(kernel = custom_kernel)
model.fit(X_iris, y_iris)
```

# Escenarios de regresión
<div style = "float:right"><a style="text-decoration:none" href = "#inicio">Inicio</a></div>

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
X = pd.DataFrame(boston.data, columns = boston.feature_names)
y = pd.Series(boston.target)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
boston.feature_names
```

```python
X
```

###### Aplicamos hold-out

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

###### Importamos la clase SVR, instanciamos el algormit y lo entrenamos

```python
from sklearn.svm import SVR
```

```python
model = SVR()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

###### Lo reentrenamos con los parámetros C = 1000 y epsilon = 1

```python
model = SVR(C=1000, epsilon=1)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

**Conclusión**

Es raro usar SVM . Tiene ventajas e inconvenientes que limitan su uso.

Ventaja principal: La maldición de la dimensionalidad. A medida que aumentamos el número de características, requerimos una gran cantidad de datos aceptable como mínimo para que el modelo puedo hacer su trabajo. Un error común es hacer ingeniería de características y aumentar mucho las características y que el dataset quede bastante pequeño. En este sentido, la principal ventaja de SVM es que funciona muy bien aún cuando el número de dimensiones es mayor que el número de muestras.

DEsventaja: EN datasets grandes, la eficiencia es pésima. Cuando los datasets tienen más de 10.000 muestras es muy complicado trabajar con ellos.
Tiende al sobreentrenamiento.
Hay que escoger bien el kernel y los parámetros del kernel.
