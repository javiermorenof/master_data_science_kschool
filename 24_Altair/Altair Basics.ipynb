```
import altair as alt
```

```
brain.columns
```

```
alt.Chart(brain).mark_tick().encode(
    x='Brain Weight'
)
```

```
alt.Chart(brain).mark_tick().encode(
    x='Body Weight'
)
```

```
# Histogram
alt.Chart(brain).mark_bar().encode(
    x=alt.X('Brain Weight',bin=True),
    y="count()"
)
```

```
hist_brain = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Brain Weight',bin=alt.Bin(maxbins=100)),
    y="count()"
).properties(
    width=300,
    height=150,
    title="Relación peso del cerebro y del cuerpo"
).interactive()
```

```
hist_body = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Body Weight',bin=alt.Bin(maxbins=100)),
    y="count()"
).properties(
    width=300,
    height=150,
    title="Relación peso del cerebro y del cuerpo"
).interactive()
```

```
scatter_brain_body = alt.Chart(brain).mark_circle().encode(
    x='Body Weight',
    y='Brain Weight'
).properties(
    width=700,
    height=300,
    title="Relación peso del cerebro y del cuerpo"
).interactive()
```

```
hist_brain = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Brain Weight',bin=alt.Bin(maxbins=100)),
    y="count()"
).properties(
    width=300,
    height=150,
    title="Relación peso del cerebro y del cuerpo"
).interactive()

hist_body = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Body Weight',bin=alt.Bin(maxbins=100)),
    y="count()"
).properties(
    width=300,
    height=150,
    title="Relación peso del cerebro y del cuerpo"
).interactive()

scatter_brain_body = alt.Chart(brain).mark_circle().encode(
    x='Body Weight',
    y='Brain Weight'
).properties(
    width=700,
    height=300,
    title="Relación peso del cerebro y del cuerpo"
).interactive()

comp_brain = (hist_brain|hist_body)&scatter_brain_body
comp_brain
```

```
trends = pd.read_csv("https://raw.githubusercontent.com/rezpe/datos_viz/master/google_trends.csv")
```

```
trends.head()
```

```
trends.dtypes
```

```
trends["search_term"].unique()
```

```
trend_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y="mean(value)",
    color="search_term"
)
```

```
trend_line.save("trend.html")
trend_line
```

```
trend_bar = alt.Chart(trends).mark_bar().encode(
    x="search_term",
    y="value",
    color="search_term"
)
```

```
trend_bar|trend_line
```

```
select_term = alt.selection(type="single",encodings=["x"])

trend_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y="mean(value)",
    color="search_term"
).transform_filter(
    select_term
)

trend_bar = alt.Chart(trends).mark_bar().encode(
    x="search_term",
    y="value",
    color="search_term",
    tooltip="search_term"
).properties(
    selection=select_term
)

trend_bar|trend_line
```

```
select_date = alt.selection(type="interval",encodings=["x"])

trend_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y="mean(value)",
    color="search_term"
).properties(
    selection=select_date
)

trend_bar = alt.Chart(trends).mark_bar().encode(
    x="search_term",
    y="value",
    color="search_term",
    tooltip="search_term"
).transform_filter(
    select_date
)

trend_bar|trend_line
```

```
select_date = alt.selection(type="interval",encodings=["x"])

trend_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y="mean(value)",
    color="search_term"
).add_selection(select_date)

trend_bar = alt.Chart(trends).mark_bar().encode(
    x="search_term",
    y="mean(value)",
    color="search_term"
).transform_filter(
    select_date
)

trend_bar|trend_line
```

```
select_date = alt.selection(type="interval",encodings=["x"])

mini_trend_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y="mean(value)",
    color="search_term"
).properties(
    height=100
).add_selection(
    select_date
)

trend_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y="mean(value)",
    color="search_term"
).transform_filter(
    select_date
)

mini_trend_line&trend_line
```

```
world_temp = pd.read_csv("https://raw.githubusercontent.com/rezpe/datos_viz/master/worldTemperatures.csv",header=None)
```

```
world_temp.columns.astype(str)
```

```
world_temp.columns=['0', '1', '2', 'date', '4', 'temp', '6', '7', '8', '9', '10', 'lat', 'lon',
       'country', 'city']
```

```
world_temp.head()
```

```
alt.data_transformers.disable_max_rows()
```

```
world_temp.to_csv("test.csv")
```

```
temp = world_temp.groupby("country").mean().reset_index()

alt.Chart(temp).mark_bar().encode(
    x=alt.X("country",sort=alt.Sort(field="temp",op="mean",order="descending")),
    y="mean(temp)"
)
```

```
temp = world_temp.groupby("city").mean().reset_index()

alt.Chart(temp).mark_bar().encode(
    x=alt.X("city",sort=alt.Sort(field="temp",op="mean",order="descending")),
    y="mean(temp)",
    tooltip="city"
).properties(
    width=600
)
```

```
world_temp["date"]=pd.to_datetime(world_temp["date"],format="%Y%m%d")
```

```
temp = world_temp.groupby("date").mean().reset_index()

alt.Chart(temp).mark_line().encode(
    x=alt.X("date:T"),
    y="mean(temp)",
).properties(
    width=600
)
```

```
select_country = alt.selection(type="single",encodings=["x"])

temp = world_temp.groupby(["country","date"]).mean().reset_index()

temp_bar = alt.Chart(temp).mark_bar().encode(
    x=alt.X("country",sort=alt.Sort(field="temp",op="mean",order="descending"),axis=None),
    y="mean(temp)",
    tooltip="country"
).properties(
    width=600
).add_selection(select_country)

temp_line = alt.Chart(temp).mark_line().encode(
    x=alt.X("date:T"),
    y="mean(temp)",
).properties(
    width=600
).transform_filter(
    select_country
)

temp_bar&temp_line
```

```
snake = pd.read_excel("https://github.com/rezpe/datos_viz/blob/master/SnakeOil_2.xlsx?raw=true")
```

```
snake.head()
```

```
alt.Chart(snake).mark_circle().encode(
    x="evidence",
    y="popularity",
    color="type"
)
```

```
select_type = alt.selection(type="single",encodings=["x"])

scatter_snake = alt.Chart(snake).mark_circle().encode(
    x="evidence",
    y="popularity",
    color="type"
).transform_filter(select_type)

bar_snake = alt.Chart(snake).mark_bar().encode(
    x="type",
    y="count()",
    color="type"
).add_selection(select_type)

scatter_snake|bar_snake
```
