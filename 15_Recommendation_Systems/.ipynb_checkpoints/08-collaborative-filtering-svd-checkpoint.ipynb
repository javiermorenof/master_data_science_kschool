## Algoritmo 2: Singular Value Decomposition (SVD)

La idea principal es la misma que la del algoritmo ALS: la factorización de la matriz users-items. El cómo lo hace es lo que difiere. 

### Matrix factorization con autovalores y autovectores (eigendecomposition)

En este caso, partimos del hecho de que una matriz cuadrada $M$, de tamaño $(n$ x $n)$, puede descomponerse con autovalores y autovectores de la siguiente forma:

$$M=V \cdot diag(\boldsymbol{\lambda}) \cdot V^{-1}$$

siendo $\boldsymbol{\lambda}$ el vector de autovalores y $V$ la matriz de autovectores.

```python
'''Ejemplo'''

import numpy as np

#Supongamos que tenemos la matriz de cambio de base:

A = np.array([[5,1],[3,3]])
A
```

```python
# Calculamos los autovalores y autovectores

np.linalg.eig(A)
```

```python
# Comprobamos que A=V · diag(lambda) · V^{-1}

V           = np.array([[ 0.70710678, -0.31622777], [ 0.70710678,  0.9486833]])
V_inv       = np.linalg.inv(V)
diag_lambda = np.diag(np.array([6., 2.]))


V.dot(diag_lambda.dot(V_inv))
```

<div class  = "alert alert-success">
    
**Warning**: La descomposición de matrices por autovalores y autovectores sólo es válida para matrices cuadradas.
</div>

### SVD

La descomposición SVD permite a una matriz no cuadrada descomponerse en el producto de tres matrices de la siguiente forma:

$$A=U \cdot D \cdot V^T$$

donde:

- $A$ es una matriz no-cuadrada $m \times n$, 
- $U$ es una matriz cuadrada $m \times m$, 
- $D$ es una matriz no-cuadrada $m \times n$ y, 
- $V$ es una matriz cuadrada $n \times n$. 
- $U$ y $V$ son matrices ortogonales; i.e. $U^{-1}=U^T$ y $V^T=V^{-1}$
- $D$ es una matriz diagonal (aunque NO-cuadrada).


Además, 

- $U$ se corresponde con los autovectores de $A \cdot A^T$,
- $V$ se corresponde con los autovectores de $A^T \cdot A$ y,
- $D$ se corresponde con los autovalores de $A \cdot A^T$ y $A^T \cdot A$, pues coinciden.


<div class  = "alert alert-info">

La idea básica es que $A$ es una matriz que puede verse como una transformación lineal. Y que dicha transformación se puede descomponer en tres subtransformaciones: 1. rotación, 2. reescalado, 3. rotación. Estos tres pasos corresponden a las tres matrices $U$, $D$ y $V$.

</div>



https://en.wikipedia.org/wiki/Singular_value_decomposition

```python
"""Ejemplo:"""

A = np.array([[7, 2], [3, 4], [5, 3]])
U, D, Vt = np.linalg.svd(A)

print(A)
print(' -- ')
print(U)
print(' -- ')
print(D)
print(' -- ')
print(Vt)
```

```python
Sigma = np.zeros((A.shape[0], A.shape[1]))
Sigma[:A.shape[1], :A.shape[1]] = np.diag(D)
print(U.dot(Sigma.dot(Vt)))
```

<div class  = "alert alert-success">
    
**Nota**: La factorización de matrices mediante autovectores/autovalores se utilizan bastante en problemas de **reducción de dimensionalidad** (como PCA, Principal Component Analysis).

</div>

```python
'''Ejemplo de reducción de dimensiones con SVD y una imagen'''

import matplotlib.pyplot as plt
from PIL import Image

# Tomamos una imagen en blanco y negro
plt.style.use('classic')
img = Image.open('kschool.png')
imggray = img.convert('LA')
imgmat = np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])

print('dimensiones originales: ' + str(imgmat.shape))

plt.figure(figsize=(3, 3))
plt.imshow(imgmat, cmap='gray')
plt.show()
```

```python
# Calculamos su factorización SVD

U, D, Vt = np.linalg.svd(imgmat)
print(imgmat.shape)
print(U.shape)
print(D.shape)
print(Vt.shape)
```

```python
# Vamos a ver que con i=5,10,15... autovalores principales ya podemos hacernos una idea 
# de cómo es la imagen original

i=5 
reconstimg = np.matrix(U[:, :i]) * np.diag(D[:i]) * np.matrix(Vt[:i, :])

plt.figure(figsize=(3, 3))
plt.imshow(reconstimg, cmap='gray')
title = "n = %s" % i
plt.title(title)
plt.show()
```

### Dataset

```python
''' Descargar el fichero pequeño
    https://grouplens.org/datasets/movielens/latest/
    y lo guardamos en una carpeta local'''

folder = '/home/eduardofernandez/Escritorio/recommender_systems/ml-latest-small'
```

```python
'''Leemos los ficheros donde están las peliculas y las votaciones de los usuarios'''

import pandas as pd

pd_ratings = pd.read_csv(folder+'/ratings.csv')
print('Number of records: ' + str(len(pd_ratings)))
print('Total of users: ' + str(pd_ratings.userId.nunique()))
pd_ratings.head()
```

```python
pd_movies = pd.read_csv(folder+'/movies.csv')
print('Number of records: ' + str(len(pd_movies)))
print('Total of movies: ' + str(pd_movies.movieId.nunique()))
pd_movies.head()
```

### Creamos la matriz users-items

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
pd_pivot = pd_ratings.pivot(index = 'userId', columns ='movieId', values = 'rating').fillna(0)
pd_pivot
```

```python
'''Para poder cargarlo en las librerías de álgebra lineal, es necesario que estén en numpy array'''

pd_pivot_numpy = pd_pivot.as_matrix()

# Normalizamos restándole la media 
# (y deberíamos dividir por la desviación típica, pero no lo hacemos porque hay demasiados ceros)
user_ratings_mean = np.mean(pd_pivot_numpy, axis = 1)
R_demeaned = pd_pivot_numpy - user_ratings_mean.reshape(-1, 1)
```

```python
"""En este ejemplo, trabajamos con la librería scipy porque la función svds nos permite reducir
   el número de dimensiones que queremos sacar automaticamente (latent factors). Así trabajamos 
   con tablas más pequeñas pero suficientemente representativas."""

from scipy.sparse.linalg import svds

U, sigma, Vt = svds(R_demeaned, k = 50)

print(U.shape)
print(sigma.shape)
print(Vt.shape)
```

```python
"""Realizamos el cálculo de U*D*VT (pero solo con los estados latentes que queramos)"""

# Recomendaciones finales
all_user_predicted_ratings = np.dot(np.dot(U, np.diag(sigma)), Vt)

# Volvemos a sumar las medias, para 'reconstruir' los valores originales
all_user_predicted_ratings = all_user_predicted_ratings + user_ratings_mean.reshape(-1, 1)

# Lo pasamos a pandas
pd_predictions = pd.DataFrame(all_user_predicted_ratings, columns = pd_pivot.columns)
pd_predictions
```

### Recomendación de películas

```python
"""Realizamos recomendaciones de películas, pero habría que depurar para hacer recomendaciones que no haya
   visto el usuario aún."""

def recommend_movies(pd_predictions, userID, movies_df, original_ratings_df, num_recommendations=5):
    
    # Predicciones del usuario
    user_row_number = userID - 1 # Como empieza en 1, bajarlo al 0
    sorted_user_predictions = pd_predictions.iloc[user_row_number].sort_values(ascending=False)
    
    user_data = pd_ratings[pd_ratings.userId == (userID)]
    user_full = user_data.merge(pd_movies, how = 'left', left_on = 'movieId', right_on = 'movieId')\
                     .sort_values(['rating'], ascending=False)
    
    # Recomendaciones (aunque podría haber visto algunas ya)
    recommendations = pd_movies.merge(pd.DataFrame(sorted_user_predictions).reset_index(), how = 'left',
           left_on = 'movieId',
           right_on = 'movieId')

    return recommendations

predictions = recommend_movies(pd_predictions, 10, pd_movies, pd_ratings, 10)
predictions
```
