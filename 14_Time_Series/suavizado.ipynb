```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn")                  # Establecemos el estilo gr치fico de seaborn por defecto
plt.rcParams["figure.figsize"] = (15, 5)  # Tama침o por defecto de las gr치ficas a crear
```

```python
candy = pd.read_csv("candy_production.csv", parse_dates = ["observation_date"])
candy.rename({"observation_date": "date", "IPG3113N": "production"}, inplace = True, axis = 1)
candy.set_index("date", inplace = True)
candy.head()
```

Para simplificar el gr치fico, nos quedamos con datos posteriores a 2009

```python
candy = candy.truncate(before = "2010")
```

```python
candy.plot()
plt.suptitle("Datos originales", fontsize = 20)
plt.show()
```

# rolling window

```python
candy_roll = candy.rolling(window = 10, min_periods = 1).mean()
```

```python
candy_roll.plot()
plt.suptitle("Datos suavizados con una rolling window", fontsize = 20)
plt.show()
```

# resample

```python
candy_res = candy.resample("3M").mean()
```

```python
candy_res.plot()
plt.suptitle("Datos suavizados tras remuestrear los datos", fontsize = 20)
plt.show()
```
